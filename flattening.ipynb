```python
import pandas as pd
import numpy as np
from  numpy import genfromtxt
import os
import re
```

```python
corr_file_path=r"C:\Users\ashis\Desktop\Correlation\nyu_1"
output_file_path= r'C:\Users\ashis\Desktop\ashi.csv'

Pear_FC=pd.DataFrame()
sub_list=[]

size_subId=5 # if the length of subID in filename is 5 put 5 here. e.g filename ROICorrelation_Sub_003.txt, here SUBID=003 
            # length of subID is 3

for root, dirs, files in os.walk(corr_file_path):
    for file in files:
        filename=(root+"/"+file).replace("\\","/") #Getting the filename
        
        SUB_ID=re.findall('[0-9]{'+str(size_subId)+'}',file)[0] # extracting subject id from file_name
        sub_list.append(SUB_ID)  # adding subject id to a list
        
        corr_file=genfromtxt(filename,delimiter='\t')[:,:-1]
        
        pear_vec=corr_file[np.triu_indices(np.shape(corr_file)[0],1)]  #fetching the upper triangular matrix and flattening
        pear_vec_row=pd.DataFrame(pd.DataFrame(pear_vec).T) # transposing the column generated to a row
        Pear_FC=Pear_FC.append(pear_vec_row) # adding the row to the dataframe
        
Pear_FC['SUB_ID'] = ''
Pear_FC['SUB_ID'] = pd.to_numeric(sub_list)
Pear_FC.to_csv(output_file_path, index=None)
```
